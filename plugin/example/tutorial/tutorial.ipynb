Here, the Jagla potential is implemented and tested for two sites with a separation dist of 1.2345, sigma of 1, cutoff of 3 and epsilon of 1.
The analytical energy is thus (3 - 1.2345)/(3 - 1) = 0.88275.

```python
import sys
import subprocess
import unittest
import pandas as pd
```

Write a two site xyz file for testing the energy with a cubic box length of 8, one particle on the origin and the second particle a distance of 1.2345 away from the origin along the z-axis.

```python
with open('example_two.xyz', 'w') as myfile: myfile.write("""2
-1 8 8 8
0 0 0 0
1 0 0 1.2345
""")
```

Prepare to run a feasst simulation with the given parameters

```python
def run_fst(params):
    with open("launch.txt", "w") as myfile: myfile.write("""
MonteCarlo
Configuration particle_type0 /feasst/forcefield/atom.fstprt xyz_file example_two.xyz
Potential Model ModelExample example_argument {example_argument}
ThermoParams beta 1
Metropolis
Log file_name log.csv max_precision true clear_file true
Run num_trials 1
""".format(**params))
    syscode = subprocess.call("../../../build/bin/fst < launch.txt > launch.log", shell=True, executable='/bin/bash')
    if syscode > 0: sys.exit(1)
```

Run the test and check the energy.

```python
class TestExampleText(unittest.TestCase):
    def test_model(self):
        run_fst({"example_argument": "0.5"})
        df = pd.read_csv('log.csv')
        self.assertEqual(2, df['p0'][0])
        self.assertEqual(0.88275, df['energy'][0])
        
unittest.main(argv=[''], verbosity=2, exit=False)
```

Similarly test the python interface.

```python
import feasst

class TestExamplePython(unittest.TestCase):
    def test_model(self):
        example = feasst.MakeModelExample({"example_argument": "0.5"})
        self.assertEqual(0, example.energy(0, 0, 0, feasst.ModelParams()))
        self.assertEqual(0.5, example.example_argument())

unittest.main(argv=[''], verbosity=2, exit=False)
```
